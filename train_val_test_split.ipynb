```python
import json
from datasets import Dataset
import os
```

```python
clean_file = "/content/gdrive/MyDrive/TelegramData/clean_dataset.json"

with open(clean_file, 'r') as f:
  data = json.load(f)
```

```python
corpus = [(post, coms) for post in data.keys() for coms in data[post]]
len(corpus)
```

```python
corpus[1]
```

```python
examples = [
    {
        # "input_text": f"Пост: {post} Тип реакции: {cluster}",
        "input_text": f"Пост: {post}",
        "output_text": comment
    }
    for post, comment in corpus
]
```

```python
dataset = Dataset.from_list(examples)

# разбиваем выборку
train_test = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = train_test["train"]
val_test = train_test["test"].train_test_split(test_size=0.5, seed=42)
val_dataset, test_dataset = val_test["train"], val_test["test"]
```

```python
val_dataset
```

```python
path_ = "/content/gdrive/MyDrive/TelegramData/prepared_data"
train_dataset.to_json(os.path.join(path_, "train.json"))
val_dataset.to_json(os.path.join(path_, "val.json"))
test_dataset.to_json(os.path.join(path_, "test.json"))
```
